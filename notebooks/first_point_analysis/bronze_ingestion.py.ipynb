```python
# Bronze ingestion: Wimbledon 2023 point-by-point data

source_url = "https://raw.githubusercontent.com/JeffSackmann/tennis_slam_pointbypoint/master/2023-wimbledon-points.csv"
volume_path = "/Volumes/workspace/default/my_data_volume"
file_name = "2023-wimbledon-points.csv"
full_path = f"{volume_path}/{file_name}"

# Create the volume if it doesn't exist
spark.sql(
    "CREATE VOLUME IF NOT EXISTS workspace.default.my_data_volume"
)

# Copy the file from the URL to the volume
dbutils.fs.cp(
    source_url,
    full_path
)

# Read the file from the volume
df_raw = (
    spark.read
    .option("header", "true")
    .option("inferSchema", "true")
    .csv(full_path)
)

display(df_raw)
```
